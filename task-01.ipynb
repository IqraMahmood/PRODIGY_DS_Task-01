```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = pd.read_csv("sub-division_population_of_pakistan.csv")
df.head()
```

```python
print("The name of provinces in pakistan is ",df['PROVINCE'].unique())
```

```python
# Now we build bar chat to see total district in each province
```

```python
divisions=df.groupby('PROVINCE')['DIVISION'].count()
divisions
```

```python
x = df.groupby('PROVINCE')['DIVISION'].nunique().index
y = df.groupby('PROVINCE')['DIVISION'].nunique().values
plt.bar(x,y)
plt.title('Divisions in each province of Pakistan')
plt.ylabel('Counts of Divisions')
```
